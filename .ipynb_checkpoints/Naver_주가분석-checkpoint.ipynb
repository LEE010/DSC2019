```python
def issues_checker(issues, population, use_range=False):
    dates = []
    topics = []
    change_rates = []
    p_values = []
    
    for issue_date,issue in issues.items():
        date = datetime.strptime(issue_date,'%Y-%m-%d')
        if use_range == False:
            
            year, month, _ = list(map(int,issue_date.split('-')))

            start = datetime(year=year,month=month,day=1)
            if month == 12:
                end = datetime(year=year+1,month=1,day=1)
            else:
                end = datetime(year=year, month=month+1,day=1)

        else:
            start = date - timedelta(days=use_range)
            end = date + timedelta(days=use_range+1)
        
        issue_data = population.loc[population.date == datetime.strptime(issue_date,'%Y-%m-%d')]
        specimen = population.loc[(population.date >= start) & (population.date < end)]    
    
        _, p = stats.ttest_1samp(specimen.change_rate,issue_data.change_rate)
        
        dates.append(date.date())
        topics.append(issue)
        try:
            change_rates.append(round(issue_data.change_rate.values[0],2))
            p_values.append(round(p[0],3))
        except IndexError:
            change_rates.append(np.nan)
            p_values.append(np.nan)
    return pd.DataFrame({'date':dates, 'topic':topics, 'change_rate':change_rates, 'p-value':p_values}).set_index('date').sort_index()
```

```python
def issues_checker_week(issues, population, use_range=False):
    dates = []
    topics = []
    change_rates = []
    p_values = []
    
    for issue_date,issue in issues.items():
        date = datetime.strptime(issue_date,'%Y-%m-%d')
        if use_range == False:
            
            year, month, _ = list(map(int,issue_date.split('-')))

            start = datetime(year=year,month=month,day=1)
            if month == 12:
                end = datetime(year=year+1,month=1,day=1)
            else:
                end = datetime(year=year, month=month+1,day=1)

        else:
            start = date - timedelta(days=use_range)
            end = date + timedelta(days=use_range+1)
        
        issue_data = population.loc[(population.date >= date)&(population.date <= date + timedelta(days=7))]
        
        specimen = population.loc[(population.date >= start) & (population.date < end)]    
    
        _, p = stats.ttest_1samp(specimen.change_rate,issue_data.change_rate.mean())
        
        dates.append(date.date())
        topics.append(issue)
#         print(p)
        try:
#             print(issue_data.change_rate.mean())
            change_rates.append(round(issue_data.change_rate.mean(),2))
            p_values.append(round(p,3))
        except IndexError:
            change_rates.append(np.nan)
            p_values.append(np.nan)
#     print(list(map(len,[dates,topics,change_rates,p_values])))
#     print(p_values)
    return pd.DataFrame({'date':dates, 'topic':topics, 'change_rate':change_rates, 'p-value':p_values}).set_index('date').sort_index()
```

```python
def issues_checker_week2(issues, population, use_range=False):
    dates = []
    topics = []
    change_rates = []
    p_values = []
    means = []
    for issue_date,issue in issues.items():
        date = datetime.strptime(issue_date,'%Y-%m-%d')
        if use_range != False:
            start = date - timedelta(days=use_range)
            end = date + timedelta(days=use_range+1)
            specimen = population.loc[(population.date >= start)&(population.date <= end)]
        else:
            year, month, _ = list(map(int,issue_date.split('-')))

            start = datetime(year=year,month=month,day=1)
            if month == 12:
                end = datetime(year=year+1,month=1,day=1)
            else:
                end = datetime(year=year, month=month+1,day=1)
            specimen = population.loc[(population.date >=start)&(population.date <=end)]
        
        issue_data = population.loc[(population.date > date)&(population.date <= date + timedelta(days=8))]
        
        if use_range == False:
            _, p = stats.ttest_ind(specimen.change_rate,issue_data.change_rate,equal_var=False)
        else:
            _, p = stats.ttest_1samp(specimen.change_rate,issue_data.change_rate.mean())
        dates.append(date.date())
        topics.append(issue)
        means.append(round(specimen.change_rate.mean(),2))
        try:
#             print(issue_data.change_rate.mean())
            change_rates.append(round(issue_data.change_rate.mean(),2))
            p_values.append(round(p,3))
        except IndexError:
            change_rates.append(np.nan)
            p_values.append(np.nan)
#     print(list(map(len,[dates,topics,change_rates,p_values])))
#     print(p_values)
#     print(means)
    return pd.DataFrame({'date':dates, 'topic':topics, 'change_rate':change_rates,'mean':means, 'p-value':p_values}).set_index('date').sort_index()
```

statsmodels.tsa.adfuller

입력 인수  
  
x : 시계열 자료  
maxlag : ADF 검정에 사용할 시차의 수. 디폴트 12*(nobs/100)^{1/4}  
regression : 검정 사용할 모형.  
'nc': 결정론적 추세 없음. DF 검정  
'c': 상수항만 사용  
'ct': 상수항과 1차 추세 사용  
'ctt': 상수항과 1차 추세, 2차 추세 사용  
autolag : 검정 AR 모형의 차수를 자동 결정하는 알고리즘 {‘AIC’, ‘BIC’, ‘t-stat’, None}  

반환값  
  
adf : 검정 통계량  
pvalue : MacKinnon(1994) 방식을 사용한 p-value 추정치  
usedlag : 사용된 시차의 수  
nobs : 분석에 자료의 수  
critical values : 1%, 5%, 10% 수준에 해당되는 검정 통계량의 값

```python
def get_forecast(series,days):
    date = series.index[-1] + timedelta(days=1)
    history = series.append(pd.Series(index=pd.date_range(date,date+timedelta(days-1))))
#     print(history[:-days-1])
    for i in range(days):
#         print(history[:-(days-i)])
        arima = ARIMA(history[:-(days-i)],order=(1,1,1)) # p, d, q
        arima_result = arima.fit(trend='c',full_output=True,disp=1)
        history[date] = arima_result.forecast()[0][0]
        date += timedelta(days=1)
    return history[-days-1:]
```

감소하거나, 유의미하지 않게 나온 결과에 대해 자세히 보겠습니다.

확인 결과 위의 그래프처럼 이전 부터 주식의 추세가 하락세에서 반등한 것을 확인할 수 있었습니다. 대부분 기사가 발표된 후 주가가 상승하였는데 통계적으로도 유의미한지 기사 발표 후 일주일동안의 주가의 평균을 이용해 확인해보겠습니다.

대부분의 이슈의 일주일 평균이 상승하는 것을 확인할 수 있지만, 유의미하지 않은 결과들이 보입니다.

이러한 결과를 보이는 토픽에 대해 살펴본 결과 파파고, 클로바처럼 결과물이 아닌 과정에 대한 기사들에 대해서 이러한 공통점을 보입니다.
