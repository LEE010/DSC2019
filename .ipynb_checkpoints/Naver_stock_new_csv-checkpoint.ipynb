```python
import numpy as np
import pandas as pd
```

```python
url = 'http://finance.naver.com/item/sise_day.nhn?code=035420'

url
```

```python
# import lxml
df_naver = pd.DataFrame()

for page in range(82,0,-1):
    pg_url = '{url}&page={page}'.format(url=url, page=page)
    df_naver = df_naver.append(pd.read_html(pg_url, header=0)[0], ignore_index=True) 
```

```python
df_naver = df_naver.rename(columns= {'날짜': 'date', '종가': 'close_val', '전일비': 'change_val', '시가': 'open_val', '고가': 'high_val', '저가': 'low_val', '거래량': 'acc_quant'})
```

```python
df_naver.head()
```

```python
df_naver['date'] = pd.to_datetime(df_naver['date'])
df_naver = df_naver.sort_values(by=['date'],ascending=True)
df_naver.head()
```

```python
df_naver.isnull().sum()
```

```python
df_naver = df_naver.dropna()
```

```python
df_naver.head()
```

```python
naver = df_naver.reset_index().drop('index',axis=1)
```

```python
naver.head()
```

```python
naver.to_csv('./data/naver_stock.csv',mode='w')
```

```python
prev_close_val = pd.Series(0).append(naver.close_val).reset_index().drop('index',axis=1)
```

```python
naver['change_val'] = naver['close_val'] - prev_close_val[0]
```

```python
naver['change_rate'] = naver['change_val']/naver['close_val'] * 100
naver.head()
```

```python
naver['change_val'] = 0
naver['change_rate'] = 0
naver.head(1)
```

```python
naver = naver.drop(0,axis=0).reset_index().drop('index',axis=1)
```

```python
naver.to_csv('./data/naver_stock_clear.csv')
```
